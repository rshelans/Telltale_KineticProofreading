```python
def findMode(track, bins=1000):
    kernal = scipy.stats.gaussian_kde(track)
    x      = scipy.arange(min(track),max(track),(max(track)-min(track))/bins)
    sample = kernal(x)
    argmax = x[scipy.argmax(sample)]
#     counts,edges = scipy.histogram(track [track < scipy.mean(track) + 2 * scipy.std(track) ]    ,bins=bins)
#     arg          = scipy.argmax(counts)
#     mode         = scipy.mean(edges[arg:arg+2])
    return(argmax)

pad = lambda x,length: scipy.pad(x,  [0, (length - len(x))  ],  mode='constant', constant_values=scipy.nan)

def moded(tracks):
    return([track - findMode(track) for track in tracks])


def correct(tracks):
    longest = max([len(track) for track in tracks])
    emd     = PyEMD.EMD()
    median  = scipy.nanmedian([pad(track,longest) for track in tracks], axis=0)
    trend   = emd.emd(median)[-1]
    ctracks = ( track/trend[:len(track)]             for track in tracks )
    ctracks = [ track - findMode(track) for track in ctracks]
    return(ctracks)    
    
def autoCorrelate(tracks):
    with multiprocessing.Pool(processes=16) as pool:
        autos  = pool.starmap(AutoCorrelate.AutoCorrelate.Trace, zip(tracks,[2.5]*len(tracks)))
        ac     = AutoCorrelate.AutoCorrelate.global_mean_global_variance(autos)
    return(ac, scipy.array(autos))   

twoStateModel = lambda t, O, k, l: O * scipy.exp(-l * t)  + (1-O) * scipy.exp(-k*t)
def fitTwoStateModel(x, y, bounds    = [0 , (1,100,100)], p0 = [0.5, 0.001, 0.001], *kargs, **kwargs):
    fit, pcov  = scipy.optimize.curve_fit(f = twoStateModel, xdata  = x, ydata  = y, p0  = p0, bounds = bounds, maxfev=1000,*kargs,**kwargs)
    k1, k2, k3 = fit
    if k2 > k3:
        return([k1  , k2, k3])
    else:
        return([1-k1, k3, k2])

def plotAC(tracks, axes = plt, *args,**kwargs):
    with multiprocessing.Pool(processes=16) as pool:
        autos  = pool.starmap(AutoCorrelate.AutoCorrelate.Trace, zip(tracks,[2.5]*len(tracks)))
        ac     = AutoCorrelate.AutoCorrelate.global_mean_global_variance(autos)
        axes.plot(ac, *args, **kwargs)
    return(ac, autos)   
    
def plotMedian(tracks, axes = plt, *args, **kwargs):
    longest = max([len(track) for track in tracks])
    median = scipy.nanmedian([pad(track,longest) for track in tracks],axis=0)
    axes.plot(median, *args, **kwargs)
    return(median)

def smooth(x, window_len=10):
    assert window_len % 2 ==0, "window length should be even Other wise output length different then input length."
    ##https://scipy-cookbook.readthedocs.io/items/SignalSmooth.html
    ##
    s = scipy.r_[x[window_len-1:0:-1],x,x[-2:-window_len-1:-1]]
    w = scipy.ones(window_len,'d')#numpy.ones(window_len,'d')#['ones', 'hanning', 'hamming', 'bartlett', 'blackman']
    return(scipy.convolve(w/w.sum(),s,mode='valid')[(window_len//2-1):-(window_len//2)])


standardBackgroundModel = lambda segment: scipy.median(segment) > 4
def classifyBreaks(bps, track, model = standardBackgroundModel):
    labels   = scipy.array([ model(track[i:j]) for i,j in zip(bps[:-1], bps[1:])]).astype(int)
    return(labels)

def mergeBreaks(brks, labels):
    dx = labels[1:] - labels[:-1]
    nbrks = scipy.concatenate([[brks[0]],  brks[scipy.nonzero(dx)[0]+1] ,[brks[-1]]])
    return(nbrks)

import ruptures
def CPD(track, min_size = 4, pen = 15):#15
    cost  = ruptures.costs.CostNormal()#   .costs.CostL2().fit(track)  #pen100# #.costs.CostL1().fit(track)
    model = ruptures.Window(width=25, custom_cost = cost, min_size =min_size).fit(track)
    model = ruptures.Pelt(custom_cost = cost, min_size=min_size).fit(track)
    bps   = scipy.array([0] + model.predict(pen=pen))
    return(bps)
```

```python
def flatten(array):
    return(scipy.array([i for d in array for i in d]))

def fitDiscreteExponentialSojurn(sojurn):
    bins    = scipy.bincount(sojurn)
    x       = scipy.arange(len(bins)) * 2.5 
    y       = scipy.cumsum(bins)/sum(bins)
    fit, cov = scipy.optimize.curve_fit(scipy.stats.expon.cdf,xdata=x,ydata=y,p0=[10,1],bounds=[(0,1),(100,500*2.5)])
    return(fit,cov)

def getDwells(dwells, labels,bins=15): 
    on   = scipy.histogram(dwells[labels]*2.5,bins=bins, range = (0,1250))[0]
    return(on/sum(on))

def getBinnedExpectedValue(x, loc, scale):
    evalue      =  scipy.stats.expon.cdf(x[1:],loc,scale) - scipy.stats.expon.cdf(x[:-1],loc,scale)
    return( evalue )



def sojurnBootstrap(breaks, tracks, nbins=18, iterations=10000, threshold = 0.25):
    BackgroundModel = lambda segment: scipy.median(segment) > threshold
    slices     = scipy.array([list(zip(bks[:-1],bks[1:])) for bks in breaks])
    dwells     = scipy.array([[(j-i) for i,j in brks] for brks in slices])
    labels     = scipy.array([[BackgroundModel(track[i:j]) for i,j in brks] for brks,track in zip(slices, tracks)])
    bootstraps = [scipy.random.randint(low=0,high=len(slices),size = len(slices)) for i in range(iterations)]
    bootondwells  = scipy.sort(scipy.array([getDwells(flatten(dwells[strap]),flatten(labels[strap]),bins=nbins) for strap in bootstraps]),axis=0)
    bootoffdwells = scipy.sort(scipy.array([getDwells(flatten(dwells[strap]),~flatten(labels[strap]),bins=nbins) for strap in bootstraps]),axis=0)
    return (dwells,labels, bootondwells, bootoffdwells, bootstraps)

iterations    =10000
cuttoff       = int(iterations * 0.05)
nbins         =scipy.arange(0,1260+60,60)
dwells102,labels102,bootondwells102,bootoffdwells102,bootstraps102 = sojurnBootstrap(breaks102, tracks102,iterations=iterations,nbins=nbins)
```

```python
def _cpd_(X, offset =0, min_segment_size=1):
    #print("_CPD_",len(X),offset)
    T   = len(X)
    if T <= 2 * min_segment_size:
        return([])
    ##ACTUAL COMPUTATION
    N   = X[-1]
    V   = scipy.arange(1,T+1)/T
    K   = X[min_segment_size:-min_segment_size]
    V   = V[min_segment_size:-min_segment_size]
    lgl = 2*K*scipy.log(K/V) + 2*(N-K)*scipy.log((N-K)/(1-V)) - 2*N*scipy.log(N) 
    ##ACTUAL COMPUTATION
    cp  = scipy.argmax(lgl) + min_segment_size
    #print("\t",len(K),cp)
    post= _cpd_(X[cp:  ] - X[cp], offset = cp+offset )
    pre = _cpd_(X[  :cp]        , offset = offset    )
    return(pre +   [[cp+offset, scipy.absolute(scipy.amax(lgl))   ]]      + post )
    #return(pre +   [cp+offset]     + post )

def cpd(track, percision = 1000, min_segment_size=1):
    T  = len(track)
    X  = scipy.cumsum(track * percision).astype(int)
    N  = X[-1]
    return(_cpd_(X))

plt.plot(track)
breaks = cpd(track)
breaks = sorted(breaks,key=lambda x: -x[1])
print(len(breaks))
for b,score in breaks:
    plt.axvline(b)
    

```
